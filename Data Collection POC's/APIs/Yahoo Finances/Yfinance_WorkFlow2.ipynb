```python
# Importing necessory libaraies
import yfinance as yf
from datetime import datetime
```

```python
# API for collecting stock data according to company using YFinance
def get_stock_data(company_name,
                   start_date):
    end_date = datetime.now()
    stock_data = yf.download(tickers=company_name, 
                             start=start_date, 
                             end=end_date, 
                             interval='1h')
    return stock_data
```

```python
# Sending input data for getting stock data
c_name = 'IBM'
s_date = '2021-01-01'

IBM_data = get_stock_data(c_name,
                          s_date)
```
